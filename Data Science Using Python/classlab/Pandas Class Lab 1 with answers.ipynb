```python
iris = pd.read_csv('https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv')
# Qn 16. Find mean sepal length
iris['sepal_length'].mean()
```

```python
# Qn 20. Find sum of sepal_length, sepal_width, petal_length, petal_width by species

iris = pd.read_csv('https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv')
iris.groupby('species').sum()
```

```python
# Qn 21. Find sum of sepal_length and mean petal_length
iris.groupby('species').agg({'sepal_length': 'sum', 'petal_length': 'mean'})
```

```python
# Qn 22. Write a function that applies different multipliers to the column sepal_length based on species.
# If species is setosa, sepal_length gets multiplied by 2, if it is virginica, multiply by 3, and if it is versicolor,
# multiply by 5. Return the result in a new column New_sepal_length.

def multiplier(row):
    if (row['species'] == 'setosa'):
        return(row['sepal_length'] * 2)
    elif (row['species'] == 'virginica'):
        return(row['sepal_length'] * 3)
    else:
        return(row['sepal_length'] * 5)

iris['New_sepal_length'] = iris.apply(multiplier,axis = 1)
iris.head()
```

```python
# Qn 23. Load ipl dataset matches. Check how many total matches were played in each month.

matches = pd.read_csv('matches.csv')

matches['date'] = pd.to_datetime(matches['date'])
matches.groupby(matches['date'].dt.month).id.count()
```

```python
# Qn 24. Find out match id and season of the two matches played in June.

matches[matches['date'].dt.month == 6]
```

```python
# Qn 25. Which team won maximum number of matches by month.

matches.groupby(['winner',matches['date'].dt.month]).id.count().unstack().idxmax()
```

```python
#Qn 26. For the given data, fill the missing dates

df = pd.DataFrame({'Day': {0: '2017-01-01', 1: '2017-01-02', 2: '2017-01-03', 3: '2017-01-10'},
 'Movie': {0: 'GreatGatsby', 1: 'TopGun', 2: 'Deadpool', 3: 'PlanetOfApes'},
 'Rating': {0: 5, 1: 5, 2: 1, 3: 2}})

#Solution 1 using reindex

df['Day'] = pd.to_datetime(df['Day'])
dates = pd.date_range(df.Day.min(), df.Day.max())
df = df.set_index('Day').reindex(dates, fill_value = 0).reset_index().rename(columns = {'index': 'Day'})
df
```

```python
#Qn 26 #Solution 2 using resample
df = pd.DataFrame({'Day': {0: '2017-01-01', 1: '2017-01-02', 2: '2017-01-03', 3: '2017-01-10'},
 'Movie': {0: 'GreatGatsby', 1: 'TopGun', 2: 'Deadpool', 3: 'PlanetOfApes'},
 'Rating': {0: 5, 1: 5, 2: 1, 3: 2}})

df['Day'] = pd.to_datetime(df['Day'])
df = df.set_index('Day').resample('D').asfreq().fillna(0).reset_index()
df
```

### String Methods

```python
#Qn 27. Clean the title column in the following dataframe to get only the alphabets
df = pd.DataFrame({'title': ['%3ABDca67$%7', '43fsbs@*54']})
df['title'] = df['title'].str.extract('([A-Za-z]+)', expand = False)
df
```

```python
#Qn 28. In the following dataframe, col contains bunch of strings separated by underscore(_), create a column called 
# country code which is the second last value in the string (eg: HK, SG etc)
df = pd.DataFrame({'col': ['20150115_Holiday_HK_Misc', '20150115_Holiday_SG_Misc']})
df['country code'] = df['col'].str.split('_').str[-2]
df
```

```python
#Qn 29. The Col2 in the following dataframe contains three sets of information. Extract that information and save to
# three columns called code, color and value.
df = pd.DataFrame({'Col1':['AA','AB','AAC'], 'Col2':['AN||Ind(0.9)','LN||RED(8.9)','RN||RED(9.8)'], 'Col3':['log1','log2','log3']})
df[['code', 'color', 'value']] = df['Col2'].str.extract('(.*?)\|\|(.*?)\((.*)\)', expand = True)
df
```

```python
#Qn 30. Create a column called user from the following dataframe which should be just the user id (eg, abc, xyz, blah24 etc)
df = pd.DataFrame({'username': ['abc@gmail.com', 'xyz@yahoo.com', 'blah24@gmail.com']})
df['user'] = df.username.str.split('@').str[0]
df
#Solution 2 using str.extract 

df = pd.DataFrame({'username': ['abc@gmail.com', 'xyz@yahoo.com', 'blah24@gmail.com']})
df['user'] = df.username.str.extract('(.*)@', expand = True)
df
```
