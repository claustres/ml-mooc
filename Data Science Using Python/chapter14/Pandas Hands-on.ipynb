### Missing Data

```python
iris = pd.read_csv('https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv')

iris.loc[iris.sepal_length > 5, 'sepal_length'] = np.nan
iris.loc[iris.sepal_width > 3.2, 'sepal_width'] = np.nan
iris.loc[iris.petal_length > 1.3, 'petal_length'] = np.nan
iris.head()
```

```python
iris.isnull().any()
```

```python
# Qn 10. Fill the missing values with 0
iris.sepal_length = iris.sepal_length.fillna(0)
iris.head()
```

```python
# Qn 11. Fill the missing values only in sepal_length with 0

iris.loc[:,'sepal_length'] = iris.loc[:,'sepal_length'].fillna(0)
iris.head()
```

```python
# Qn 12. Fill the missing values with non-NaN previous values

iris = pd.read_csv('https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv')
iris.loc[iris.sepal_length > 5, 'sepal_length'] = np.nan
iris.loc[iris.sepal_width > 3.2, 'sepal_width'] = np.nan
iris.loc[iris.petal_length > 1.3, 'petal_length'] = np.nan

iris.ffill().head()
```

```python
# Qn 13. Fill the missing values in all columns with sepal with non-NaN previous values and all columns with 
# petal with next non-NaN value 
iris[['sepal_length', 'sepal_width']] = iris[['sepal_length', 'sepal_width']].ffill()
iris.head()
```

```python
iris['petal_length'] = iris['petal_length'].bfill()
iris.head()
```

```python
# Qn 14. Replace species names using dictionary.
d = {'setosa': 'se', 'virginica': 'vi', 'versicolor': 've'}
iris = iris.replace(d).head()
```
