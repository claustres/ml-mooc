```python
# To subcategorize the fuelType as there is no sufficint training data for fuel type other than deiseal and benzin so, 
# categorizingin the fuel in three categories benzamin, diesel, and Other_fuel.
dfu_autos = pd.DataFrame(autr)
#dfu_autos = pd.DataFrame(autos_nw)
def check_fueltype(row):
    if row['fuelType']=='benzin':
        return 'benzin'
    elif row['fuelType']=='diesel':
        return 'diesel'
    else:
        return 'Other_fuel'
dfu_autos['check_fuel'] = dfu_autos.apply(check_fueltype, axis=1)
dfu_autos.head(10)
```

```python
# To subcategorize "year of registration" if it is required later in algo..while fine tuning the algo to imprtove accuracy..
#df_yr = pd.DataFrame(autos_nw)
df_yr = pd.DataFrame(autr)
def check_yr_reg(row):
    if row['yearOfRegistration']<=1960:
        return 'year_A'
    elif (row['yearOfRegistration']>1960)&(row['yearOfRegistration']<=1990):
        return 'year_B'
    elif row['yearOfRegistration']>1990:
        return 'year_C'
df_yr['Mod_yr_reg']= df_yr.apply(check_yr_reg, axis=1)
df_yr.head(10)
```

```python
# To subcategorize "power PS" feature vector if it is required later in algo..while fine tuning the algo to imprtove accuracy..
df_pwr = pd.DataFrame(autr)
#df_pwr = pd.DataFrame(autos_nw)
def check_pwr(row):
    if row['powerPS']<=200:
        return 'power_A'
    elif (row['powerPS']>200)&(row['powerPS']<=400):
        return 'power_B'
    elif row['powerPS']>400:
        return 'power_C'
df_yr['Mod_power']= df_pwr.apply(check_pwr, axis=1)
df_yr.head(10)
```

```python
# To subcategorize "month of Registration" if it is required later in algo..while fine tuning the algo to imprtove accuracy..
df_mr = pd.DataFrame(autr)
#df_mr = pd.DataFrame(autos_nw)
def check_mr_reg(row):
    if row['monthOfRegistration']<=3:
        return 'month_A'
    elif (row['monthOfRegistration']>3)&(row['monthOfRegistration']<=8):
        return 'month_B'
    elif row['monthOfRegistration']>8:
        return 'month_C'
df_yr['Mod_mt_reg']= df_mr.apply(check_mr_reg, axis=1)
df_yr.head(10)
```

```python
# To subcategorize "kilometer" feature vector if it is required later in algo..while fine tuning the algo to imprtove accuracy..
df_km = pd.DataFrame(autr)
#df_km = pd.DataFrame(autos_nw)
def check_km(row):
    if row['kilometer']==150000:
        return 'km_A'
    elif (row['kilometer']==100000)|(row['kilometer']==125000):
        return 'km_B'
    elif row['kilometer']<100000:
        return 'km_C'
df_yr['Mod_km']= df_km.apply(check_km, axis=1)
df_yr.head(10)
```

```python
# To subcategorize "brand" feature vector if it is required later in algo..while fine tuning the algo to imprtove accuracy..
df_br = pd.DataFrame(autr)
#df_br = pd.DataFrame(autos_nw)
def check_br(row):
    if (row['brand']=='volkswagen')|(row['brand']=='opel'):
        return 'brand_A'
    elif (row['brand']=='bmw')|(row['brand']=='mercedes_benz'):
        return 'brand_B'
    else:
        return 'brand_C'
df_br['Mod_br']= df_br.apply(check_br, axis=1)
df_br.head(10)
```

```python
# To subcategorize "vehicle type" if it is required later in algo..while fine tuning the algo to imprtove accuracy..
df_vcl_typ = pd.DataFrame(autr)
#df_vcl_typ = pd.DataFrame(autos_nw)
def check_vcl_typ(row):
    if row['vehicleType']=='limousine':
        return 'Vhcl_A'
    elif row['vehicleType']=='kleinwagen':
        return 'Vhcl_B'
    elif (row['vehicleType']=='kombi')|(row['vehicleType']=='bus'):
        return 'Vhcl_C'
    else:
        return 'Vhcl_D'
df_vcl_typ['Mod_vecl_typ']= df_vcl_typ.apply(check_vcl_typ, axis=1)
df_vcl_typ.head(10)
```

```python
# To check parameters of model..
# Techinique-1: It is observed rmse and r2 value is low using basic Multiple regressor model..
from sklearn.metrics import mean_squared_error
from math import sqrt

rmse = sqrt(mean_squared_error(y_test, y_pred))

from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)

adj_r2 = 1 - float(len(y)-1)/(len(y)-len(mlr.coef_)-1)*(1 - r2)

rmse, r2, adj_r2, mlr.coef_, mlr.intercept_
```
